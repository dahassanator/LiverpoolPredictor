```python
import pandas as pd
matches = pd.read_html('https://fbref.com/en/squads/822bd0ba/2023-2024/matchlogs/c9/schedule/Liverpool-Scores-and-Fixtures-Premier-League')
matches = matches[0]
```

```python
matches.head()
```

```python
shooting = pd.read_html('https://fbref.com/en/squads/822bd0ba/2023-2024/matchlogs/c9/shooting/Liverpool-Match-Logs-Premier-League')
shooting = shooting[0]
shooting.columns = shooting.columns.droplevel()
shooting.head()
```

```python
team_data = matches.merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt", "xG"]], on="Date")
team_data
```

```python
team_data.to_csv("liverpool_data.csv")
```
